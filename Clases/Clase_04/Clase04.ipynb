# Instalación
+ Activar tu entorno virtual de trabajo.
+ Mediante conda, instalar `jupyter`.
+ Para poder utilizar en jupyter las librerias puntuales presentes en el entorno virtual, debemos crear un *kernel*.       
      
        Un Kernel es un programa que ejecuta y examina nuestro código.
        El kernel, al igual que el entorno virtual esta creado para cumplir los requisitos de cada proyecto, por lo que se recomienda que cada entorno tenga su propio kernel.

+ Tienes que instalar `ipykernel` en tu entorno.
+ Una vez instalada la herramienta ipykernel en tu entorno, se procede a crear el kernel para tu proyecto.
+ Mediante el siguiente comando:

        python -m ipykernel install --user --name=<nombre de kernel>
        
+ Una vez realizados estos pasos, ya te encuentras en condiciones de abrir la aplicación jupyter.
+ Ejecuta la siguiente línea:

        jupyter notebook

+ Esto abrirá tu navegador web predeterminado, con la aplicación en funcionamiento.

```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
A=3
freq= 15
T=2
t=np.linspace(0,T,1000)
sig=A*np.cos(2*np.pi*freq*t)
```

```python
fig,ax = plt.subplots(figsize=(9,4), tight_layout=True)
ax.plot(t,sig)
ax.set_title('Onda Coseno')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Amplitud')
```
