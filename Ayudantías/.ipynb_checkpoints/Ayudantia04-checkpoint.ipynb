```python
import numpy as np
from matplotlib import pyplot as plt 
%matplotlib notebook
import librosa as lib
from librosa import display as libdis
import IPython.display as disp
```

```python
sig, fs = lib.load('ref.wav',sr=44100,duration=9)
```

```python
sig=sig/max(sig)
sig1 = sig[0:5*fs]
dur = len(sig1)/fs
t = np.linspace(0,dur,int(dur*fs))
```

```python
total_fade_out = np.exp(-t)/max(np.exp(-t))
```

```python
fig1, ax1 = plt.subplots(2,2,figsize=(7,5),tight_layout=True)

#onda temporal
ax1[0,0].set_title('Onda temporal')
ax1[0,0].plot(t,sig1)
ax1[0,0].set_xlim([0,dur])

#espectro de la onda temporal
hop_length=1024
D1 = lib.amplitude_to_db(np.abs(lib.stft(sig1,hop_length=hop_length)),ref=np.max)
libdis.specshow(D1,y_axis='log', sr=fs,hop_length=hop_length,ax=ax1[0,1])
ax1[0,1].set_title('Espectrograma')
#señal con fade out
ax1[1,0].plot(t,sig1*total_fade_out)
ax1[1,0].set_xlim([0,dur])
ax1[1,0].set_xlabel('Tiempo [segundos]')

#espectrograma con fade out
D2 = lib.amplitude_to_db(np.abs(lib.stft(sig1*total_fade_out,hop_length=hop_length)),ref=np.max)
libdis.specshow(D2,y_axis='log', sr=fs,hop_length=hop_length,ax=ax1[1,1])
ax1[1,1].set_xlabel('Tiempo [segundos]')
```

```python
disp.Audio(data=sig1,rate=fs)
```

```python
disp.Audio(data=sig1*total_fade_out,rate=fs)
```

```python
t_f = np.linspace(0,2,2*fs)
t_F = np.linspace(0,9,len(sig))
```

```python
fade_in = np.exp(t_f)/max(np.exp(t_f))
fade_out = np.exp(-t_f)/max(np.exp(-t_f))
continuo = np.ones(5*fs)
fade = np.hstack((fade_in,continuo,fade_out))
```

```python
fig2, ax2 = plt.subplots(1,3,figsize=(14,5),tight_layout=True)
ax2[0].plot(t_F,sig,'b')
ax2[0].set_xlim([0,9])
ax2[0].set_ylabel('Amplitud')
ax2[0].set_xlabel('Tiempo [segundos]')
ax2[1].plot(t_F,fade,'r')
ax2[1].set_xlim([0,9])
ax2[1].set_xlabel('Tiempo [segundos]')
ax2[2].plot(t_F,sig*fade,'g')
ax2[2].set_xlim([0,9])
ax2[2].set_xlabel('Tiempo [segundos]')
```
